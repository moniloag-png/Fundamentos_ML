```python
from sklearn import preprocessing
import numpy as np
```

```python
# Carga de datos.
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Limpieza de datos: estandarización.
standardizer = preprocessing.StandardScaler()

# Estandarizamos los datos de entrenamiento
X_train_std=standardizer.fit_transform(X_train)
print(X_train_std)
```

```python
X_train_2 = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Aplicamos los métodos "fit" y "transform" por separado
f=preprocessing.StandardScaler()
f=f.fit(X_train_2)
X_train_2_std = f.transform(X_train_2)
print(X_train_2_std)
```

#### Comprobar si las matrices estandarizadas son iguales

```python
# Comprobación de igualdad absoluta
son_exactamente_iguales = np.array_equal(X_train_std, X_train_2_std)

# Comprobación considerando pequeñas variaciones decimales (Recomendado)
son_similares = np.allclose(X_train_std, X_train_2_std)

print(f"¿Son exactamente iguales?: {son_exactamente_iguales}")
print(f"¿Son iguales (tolerancia decimal)?: {son_similares}")
```

```python
# Una forma visual de comprobarlo:
# Si restas las dos matrices y el resultado es una matriz llena de ceros, son iguales:

print(X_train_std - X_train_2_std)
```
