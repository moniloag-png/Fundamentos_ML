```python
import numpy as np
from sklearn import datasets
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
# Carga de datos
heart_disease = pd.read_csv('dataset/heart.csv')

# 2. Definimos manualmente los nombres de las categorías del target
target_names = ['0 (Sano)', '1 (Enfermo)']

# 3. Imprimimos las etiquetas correspondientes al target
print(target_names)
```

```python
# Mostrar características de la tabla de datos.
# Nota: En Pandas, .shape[1] incluye el target, por lo que restamos 1 para contar solo los atributos/características.
print("Tabla de datos: %d instancias y %d atributos" % (heart_disease.shape[0], heart_disease.shape[1] - 1))
print("Valores de la clase:", set(heart_disease['target']))

# Cuantificamos el número de instancias que contiene el dataset por clase
valores, ocurrencias = np.unique(heart_disease['target'], return_counts=True)
print("Clases:", valores, "Ocurrencias por clase:", ocurrencias)
```

```python
# Test: hold-out split 80-20%. # Partición externa

# Dividimos usando la sintaxis de Pandas para separar las características de la columna 'target'
X_training, X_test, y_training, y_test = train_test_split(heart_disease.drop(columns=['target']), heart_disease['target'], test_size=0.2, 
    random_state=42,
    stratify=heart_disease['target']  # Recomendado para mantener la proporción de clases
)

# Cuantificamos las ocurrencias en el conjunto de prueba
valores_test, ocur_test = np.unique(y_test, return_counts=True)
print('Test: ', 'clases:', valores_test, ' ocurrencias: ', ocur_test)
```

```python
# Estandarizar las características de entrenamiento y de test
standardizer = StandardScaler()
X_training = standardizer.fit_transform(X_training)
X_test = standardizer.transform(X_test)
```

```python
# Validación: hold-out split 80-20%. # Partición interna
X_train, X_val, y_train, y_val = train_test_split(X_training, y_training, test_size=0.2, random_state=42)
valores_train, ocur_train = np.unique(y_train, return_counts=True)
print('Entrenamiento: ', ' clases:', valores_train, '  ocurrencias:', ocur_train)

valores_val, ocur_val = np.unique(y_val, return_counts=True)
print('Validation:    ', ' clases:', valores_val, '  ocurrencias:', ocur_val)
```

```python
# Construcción del objeto que contiene el algoritmo de aprendizaje.
clf = DummyClassifier(strategy='prior', random_state=42)
```

```python
# Entrenamiento del algoritmo de aprendizaje.
clf = clf.fit(X_train, y_train)
```

```python
# Evaluación del algoritmo de aprendizaje con el método "score" que devuelve directamente la métrica de 'accuracy'
val_accuracy = clf.score(X_val, y_val)
print("Exactitud en validación: ", val_accuracy)

test_accuracy = clf.score(X_test, y_test)
print("Exactitud en test: ", test_accuracy)
```

```python
# Obtenemos las predicciones sobre conjunto de validación y de test
y_pred_val = clf.predict(X_val)
print('Predicciones de validación ', y_pred_val)
print('Etiquetas reales validación', y_val)

y_pred_test = clf.predict(X_test)
print('\nPredicciones de test ', y_pred_val)
print('Etiquetas reales test', y_val)
```

```python
# Aplicamos un ejemplo con un clasificador más complejo que el "dummyclassifier"
from sklearn.svm import SVC
svc = SVC(C=0.5) # Definimos algoritmo
svc.fit(X_train, y_train) # Entrenamos modelo

val_accuracy = svc.score(X_val, y_val) # Evaluamos modelo en validación
print('Exactitud en validación ', np.round(val_accuracy*100, 4), '%')

test_accuracy = svc.score(X_test, y_test) # Evaluamos modelo en test
print('Exactitud en test ', np.round(test_accuracy*100, 4), '%')
```

```python
# Guardar modelo
import pickle
with open('models/model.pkl', 'wb') as fw:
    pickle.dump(svc, fw)

# Cargar modelo
with open('models/model.pkl', 'rb') as fr:
    modelo=pickle.load(fr)
print("Modelo cargado")

#Guardamos el escalado
with open('models/scaler.pkl', 'wb') as fw:
     pickle.dump(standardizer, fw)

#Guardamos el scaler
with open('models/scaler.pkl', 'rb') as f:
   scaler_cargado=pickle.load(f)

print("Modelo y scaler cargados")
```

#### **Probamos:** Ingresar los nuevos datos por teclado y generar la predicción

```python
# Ingresar los nuevos parámetros médicos por teclado para la predicción de riesgo cardíaco

print("Ingrese los datos clínicos del paciente para la predicción:")

# Variables numéricas continuas
age = float(input("Edad (en años, ej. 52): "))
trestbps = float(input("Presión arterial en reposo en mm Hg (ej. 125): "))
chol = float(input("Colesterol sérico en mg/dl (ej. 212): "))
thalach = float(input("Frecuencia cardíaca máxima alcanzada (ej. 168): "))
oldpeak = float(input("Depresión del segmento ST inducida por el ejercicio (ej. 1.0): "))

# Variables categóricas y binarias (codificadas numéricamente como el dataset original)
sex = int(input("Sexo (1 = masculino, 0 = femenino): "))
cp = int(input("Tipo de dolor torácico (0 a 3): "))
fbs = int(input("Azúcar en sangre en ayunas > 120 mg/dl (1 = verdadero, 0 = falso): "))
restecg = int(input("Resultados electrocardiográficos en reposo (0, 1 o 2): "))
exang = int(input("Angina inducida por el ejercicio (1 = sí, 0 = no): "))
slope = int(input("Pendiente del segmento ST en el ejercicio máximo (0, 1 o 2): "))
ca = int(input("Número de vasos principales coloreados por fluoroscopia (0 a 4): "))
thal = int(input("Resultados de la prueba Thalassemia (0 = normal, 1 = defecto fijo, 2 = defecto reversible, 3 = no especificado): "))
```

```python
#Crear la lista con los nuevos datos
nuevo_paciente = [[age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal]]

#Escalar los datos
# Es fundamental usar el mismo scaler con el que se entrenó el modelo
nuevo_paciente_escalado = scaler_cargado.transform(nuevo_paciente)
print(nuevo_paciente_escalado)
```

```python
#Predicción

prediccion = modelo.predict(nuevo_paciente)
print(f"Resultado de la predicción (clase): {prediccion[0]}")

#Condicional para mostrar el nombre de la especie
# Nota: prediccion[0] se usa para obtener el valor numérico del array
if prediccion[0] == 0:
    print("Diagnóstico: El paciente se encuentra SANO (Ausencia de enfermedad cardíaca).")
elif prediccion[0] == 1:
    print("Diagnóstico: El paciente presenta RIESGO CARDIOVASCULAR (Presencia de enfermedad cardíaca).")
```
