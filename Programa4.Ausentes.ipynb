```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
# Carga de datos
X_train = [[1,      4], 
           [np.nan, 3], 
           [7,      8]]
print(X_train)
```

```python
# Limpieza de datos: imputación valores ausentes (modelo)
imputer = SimpleImputer(strategy='mean') # Utilizamos la estrategia de la media

# Aplicar los métodos "fit" y "transform" para imputar los valores pérdidos de X_train
# la media es de cada columna, ejemplo: 1 y 7, la media es 4 ((7+1)/4)
X_train_clean = imputer.fit_transform(X_train)
print(X_train_clean)
```

```python
# Imputación de valores ausentes sobre el conjunto de test
X_test = [[np.nan, 10], 
          [6,      np.nan], 
          [8,      2]]

# Imputar los datos del conjunto de test
X_test_clean=imputer.transform(X_test)
print(X_test_clean)
```

en este caso como ya se generó el modelo, en el ejercicio de imutación en la columna 1 reemplaza el valor (np.nan) por la media (4) y en la columna 2 reemplaza el valor por la media aritmética de ((4+3+8)/3)
