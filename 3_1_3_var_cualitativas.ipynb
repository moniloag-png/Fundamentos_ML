```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
#Crear una variable de tipo lista vacia
nivel_estudios = []
#Crear o ingresar a mi lista la variable secuariada  y creo 110 veces secundaria
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras de estudio')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

# Tarea implementar con pandas esta información 

```python

import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear una Serie de Pandas con los datos originales
df_estudios = pd.Series(nivel_estudios)

# 2. Calcular la tabla de frecuencias absolutas
# (Pandas ordena automáticamente de forma descendente por frecuencia, 
# pero podemos ordenarlo alfabéticamente con .sort_index() para igualar el comportamiento de numpy.unique)
frecuencias = df_estudios.value_counts().sort_index()

print("Frecuencias con Pandas:")
print(frecuencias)

# 3. DIAGRAMA DE BARRAS CON PANDAS
plt.figure(figsize=(8, 5))
frecuencias.plot(kind='bar', color='tab:blue', edgecolor='black')
plt.title('Diagrama de barras (Pandas)')
plt.xlabel('Nivel de estudios')
plt.ylabel('Frecuencia absoluta')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 4. DIAGRAMA DE SECTORES CON PANDAS
# Definimos la lista de colores en el orden correspondiente a los índices ordenados
colores_lista = ['orange', 'green', 'blue', 'cyan', 'red'] 

plt.figure(figsize=(6, 6))
frecuencias.plot(
    kind='pie', 
    colors=colores_lista, 
    autopct='%1.2f%%', 
    startangle=180 # Ajusta el ángulo si deseas replicar la orientación exacta
)
plt.title('Diagrama de sectores (Pandas)')
plt.ylabel('') # Elimina la etiqueta lateral por defecto de la serie
plt.show()
```
